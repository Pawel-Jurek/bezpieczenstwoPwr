```python
import pandas as pd

mouse_human = pd.read_csv("dataMouse.csv")
keyboard_train = pd.read_csv("trainKeyboard.csv")
keyboard_test = pd.read_csv("testKeyboard.csv")
```

```python
columns_df1 = set(keyboard_train.columns)
columns_df2 = set(keyboard_test.columns)

only_in_df1 = columns_df1 - columns_df2
only_in_df2 = columns_df2 - columns_df1
common_columns = columns_df1 & columns_df2

print("Kolumny tylko w df1:", only_in_df1)
print("Kolumny tylko w df2:", only_in_df2)
print("Wspólne kolumny:", common_columns)
```

```python
keyboard_train.drop("user", axis='columns', inplace=True)

keyboard_human = pd.concat([keyboard_train, keyboard_test], ignore_index=True)

keyboard_bot = pd.read_csv("generating_bot_data\\bot_keyboard_events.csv")
keyboard_bot["is_bot"] = 1
keyboard_human["is_bot"] = 0

keyboard_data = pd.concat([keyboard_bot, keyboard_human], ignore_index=True)

keyboard_data.to_csv("keyboard_combined.csv", index=False)
```

```python
mouse_bot = pd.read_csv("generating_bot_data\\bot_mouse_events.csv")

mouse_bot["is_bot"] = 1
mouse_human["is_bot"] = 0

mouse_human.drop("uid", axis='columns', inplace=True)

mouse_bot.rename(columns={"x": "screen_x"}, inplace=True)
mouse_bot.rename(columns={"y": "screen_y"}, inplace=True)

# Połączenie danych
mouse_data = pd.concat([mouse_bot, mouse_human], ignore_index=True)

mouse_data.to_csv("mouse_combined.csv", index=False)
```
